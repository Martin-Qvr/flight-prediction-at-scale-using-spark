## Cleaning flights

```
rdd_flights = rdd_flights.map(lambda x: x.split(','))

## We first need to select the columns we are interested in :

rdd_flights = rdd_flights.map(lambda x: x[:7] + x[9:12])

#Delete first row and turn first column into Datetime

import datetime
from datetime import datetime 

header = rdd_flights.first()

rdd_flights = rdd_flights.filter(lambda line: line != header)
rdd_flights = rdd_flights.map(lambda x: [datetime.strptime(x[0], '%Y-%m-%d')] + x[1:])

rdd_flights.take(5)
```

## Cleaning of the weather

```
# Loading the weather Dataset

file_location = "/FileStore/tables/weather/201402hourly.txt"

rdd_weather = sc.textFile(file_location)
rdd_weather.take(5)
```

```
rdd_weather = rdd_weather.map(lambda x: x.split(','))

header = rdd_weather.first()
rdd_weather = rdd_weather.filter(lambda line: line != header)

rdd_weather = rdd_weather.map(lambda x: [x[0]] + x[1:])
rdd_weather.take(5)

## Take only the columns we need 

rdd_weather = rdd_weather.map(lambda x: x[0:3] + [x[4]] + [x[6]] + [x[8]] + [x[10]] + [x[22]] + [x[24]] + [x[26]] +[x[30]])

rdd_weather = rdd_weather.map(lambda x: (int(x[0]), x[1:]))

weather_join = rdd_weather.collect()

weather_join
```

```
weather_rdd = sc.parallelize(weather_join)
weather_rdd.take(5)
```

```
file_location = "/FileStore/tables/wban_airport_timezone.csv"
file_type = "csv"

rdd_wban = sc.textFile(file_location)
rdd_wban.take(5)
```

```
rdd_wban_2 = rdd_wban

header = rdd_wban_2.first()
rdd_wban_2 = rdd_wban_2.filter(lambda line: line != header)
rdd_wban_2 = rdd_wban_2.map(lambda x: x.split(','))

rdd_wban_2.take(5)

# We need to put the wban on first column to join on it later 
rdd_wban_2 = rdd_wban_2.map(lambda x: (int(x[1]), [x[0], x[2]]))
rdd_wban_2.take(5)

weather_rdd = weather_rdd.join(rdd_wban_2)

weather_rdd.take(5)
```

```
# Flatten the list:

weather_rdd = weather_rdd.map(lambda x: [x[0]] + x[1][0] + x[1][1])
weather_rdd.take(5)
```

```
## Let's convert our date and time into a timestamp 
import datetime 
from datetime import datetime 

weather_rdd = weather_rdd.map(lambda x: [datetime.strptime(x[1] + x[2], '%Y%m%d%H%M')] + [x[0]] + x[3:])

weather_rdd.take(5)
```

```
## Let's add Datetime_rounded column to the weather dataset. 

from datetime import timedelta 

def round_to_nearest_hour(timestamp):
    return (timestamp.replace(second=0, microsecond=0, minute=0, hour=timestamp.hour) + timedelta(hours=timestamp.minute//30)) 

weather_rdd = weather_rdd.map(lambda x: [round_to_nearest_hour(x[0])] + x[1:])

weather_rdd.take(5)
```

```
# we need to prepare the flights dataset for the join operation

list_flights = rdd_flights.collect()

rdd_flights_2 = sc.parallelize(list_flights)

rdd_join = rdd_flights_2.map(lambda x: [x[0]] + [x[3:]])

format = 'yyyy-MM-ddHHmm'

rdd_join = rdd_join.map(lambda x: [x[0] + timedelta(hours=int(x[1][2][1:3]), minutes=int(x[1][2][3:5]))] + x[1][1:3] + x[1][4:])

# we need to create the 4 columns 'SCHEDULED_DEP_TIME ACTUAL_DEP_TIME SCHEDULED_ARR_TIME ACTUAL_ARR_TIME', and also deal with the '' in the ELS_time :

def check_empty_late(x):
    if x =='':
        return 0
    else:
        return x
    
rdd_join = rdd_join.map(lambda x: [x[0]] + [x[0] + timedelta(minutes=int(check_empty_late(x[3].split('.')[0])))] + [x[0] + timedelta(minutes=int(check_empty_late(x[4].split('.')[0])))] + [x[0] + timedelta(minutes=int(check_empty_late(x[3].split('.')[0]))) + timedelta(minutes=int(check_empty_late(x[4].split('.')[0])))] + x[1:3] + x[5:])


# Add on time according to a chosen threshold 
threshold = 15 

rdd_join = rdd_join.map(lambda x: x + [0 if ((x[3] - x[2]).total_seconds() < (threshold * 60)) else 1])


# Round the flight departure time

rdd_join = rdd_join.map(lambda x: [round_to_nearest_hour(x[0])] + x[0:])

rdd_join.collect()

rdd_join.take(5)
```
